matchlist by summoner puuid

```python
ml = pd.DataFrame({'game': [],'top_b_champ': [],'jgl_b_champ': [],'mid_b_champ': [],'mid_b_dmgtaken': [],'adc_b_champ': [],'supp_b_champ': [],'supp_b_vs': [],'b_win': [],'top_r_champ': [],'jgl_r_champ': [],'mid_r_champ': [],'mid_r_dmgtaken': [],'adc_r_champ': [],'supp_r_champ': [],'supp_r_vs': [],'r_win': []})
```

```python
import json
for iterador in range(300):
  puuid = summoner.puuid[iterador]
  endpoint = f"https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count=40&{API_KEY}"
  response = requests.get(endpoint, headers=headers)
  if response.status_code == 200:
      matchlist = response.json()
      new_games = pd.DataFrame(matchlist, columns = ["game"])
      ml = ml.append(new_games, ignore_index=True)
  else:
      print(f"Error retrieving matchlist: {response.status_code}")
  time.sleep(1.5)

ml = ml.drop_duplicates()
```

match info -> tenemso los datso para al vision score

```python
import json
import time
for iterador  in range(2800,len(ml)):
  MATCH = ml.game[iterador]
  endpoint = f"https://europe.api.riotgames.com/lol/match/v5/matches/{MATCH}?{API_KEY}"
  response = requests.get(endpoint, headers=headers)
  if response.status_code == 200:
      matchlist = response.json()
      info = matchlist.get("info")
      participants =  info.get("participants")
      #
      top_azul = participants[0]
      jgl_azul = participants[1]
      mid_azul = participants[2]
      adc_azul = participants[3]
      supp_azul = participants[4]
      top_rojo = participants[5]
      jgl_rojo = participants[6]
      mid_rojo = participants[7]
      adc_rojo = participants[8]
      supp_rojo = participants[9]

      ml.loc[iterador, 'top_b_champ'] = top_azul.get("championName")
      ml.loc[iterador, 'jgl_b_champ'] = jgl_azul.get("championName")
      ml.loc[iterador, 'mid_b_champ'] = mid_azul.get("championName")
      ml.loc[iterador, 'mid_b_dmgtaken'] = mid_azul.get("totalDamageTaken")
      ml.loc[iterador, 'adc_b_champ'] = adc_azul.get("championName")
      ml.loc[iterador, 'supp_b_champ'] = supp_azul.get("championName")
      ml.loc[iterador, 'supp_b_vs'] = supp_azul.get("visionScore")
      ml.loc[iterador, 'b_win'] = supp_azul.get("win")

      ml.loc[iterador, 'top_r_champ'] = top_rojo.get("championName")
      ml.loc[iterador, 'jgl_r_champ'] = jgl_rojo.get("championName")
      ml.loc[iterador, 'mid_r_champ'] = mid_rojo.get("championName")
      ml.loc[iterador, 'mid_r_dmgtaken'] = mid_rojo.get("totalDamageTaken")
      ml.loc[iterador, 'adc_r_champ'] = adc_rojo.get("championName")
      ml.loc[iterador, 'supp_r_champ'] = supp_rojo.get("championName")
      ml.loc[iterador, 'supp_r_champ'] = supp_rojo.get("championName")
      ml.loc[iterador, 'supp_r_vs'] = supp_rojo.get("visionScore")
      ml.loc[iterador, 'r_win'] = supp_rojo.get("win")
  else:
      print(f"Error retrieving matchlist: {response.status_code}")
  time.sleep(1.5)

print(ml)
```
